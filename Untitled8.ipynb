# Is there a significant difference in salaries between males and females? Provide statistical evidence.

```python
import pandas as pd
from scipy.stats import ttest_ind

df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')

df.dropna(subset=['Salary'], inplace=True)

male_salaries = df[df['Gender'] == 'Male']['Salary']
female_salaries = df[df['Gender'] == 'Female']['Salary']

t_stat, p_value = ttest_ind(male_salaries, female_salaries)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("There is a significant difference in salaries between males and females.")
else:
    print("There is no significant difference in salaries between males and females.")
```

# How does having a personal loan affect the total combined salary of the individual and their partner?

```python
df.columns = df.columns.str.strip()


df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
df['Partner_salary'] = pd.to_numeric(df['Partner_salary'], errors='coerce')


df.dropna(subset=['Salary', 'Partner_salary', 'Personal_loan'], inplace=True)


df['Total_Combined_Salary'] = df['Salary'] + df['Partner_salary']


avg_combined_salary_by_loan_status = df.groupby('Personal_loan')['Total_Combined_Salary'].mean()

print("Average Total Combined Salary by Personal Loan Status:")
print(avg_combined_salary_by_loan_status)


import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x=avg_combined_salary_by_loan_status.index, y=avg_combined_salary_by_loan_status.values)
plt.xlabel('Personal Loan Status')
plt.ylabel('Average Total Combined Salary')
plt.title('Average Total Combined Salary by Personal Loan Status')
plt.show()
```

# Create a histogram showing the distribution of total combined salaries. Identify and discuss any skewness or outliers in the data.

```python
df.columns = df.columns.str.strip()

# Ensure relevant columns are numeric
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
df['Partner_salary'] = pd.to_numeric(df['Partner_salary'], errors='coerce')

# Handle missing values (drop rows with NaNs in relevant columns)
df.dropna(subset=['Salary', 'Partner_salary'], inplace=True)

# Calculate total combined salary
df['Total_Combined_Salary'] = df['Salary'] + df['Partner_salary']

# Plot the histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Total_Combined_Salary'], bins=30, kde=True)
plt.xlabel('Total Combined Salary')
plt.ylabel('Frequency')
plt.title('Distribution of Total Combined Salaries')
plt.show()

# Analyze skewness
skewness = df['Total_Combined_Salary'].skew()
print(f"Skewness: {skewness}")

# Identify outliers using the IQR method
Q1 = df['Total_Combined_Salary'].quantile(0.25)
Q3 = df['Total_Combined_Salary'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['Total_Combined_Salary'] < lower_bound) | (df['Total_Combined_Salary'] > upper_bound)]

print(f"Number of outliers: {len(outliers)}")
print("Outliers:")
print(outliers[['Salary', 'Partner_salary', 'Total_Combined_Salary']])
```
